```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
data =pd.read_csv("datasets.csv")
data.head()
```

```python

data.info()
print("  ")
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.shape
```

```python
data.duplicated().sum()
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python
data.dtypes
```

```python
data['id'] = data['id'].astype(object)
data.dtypes
```

```python
df = data[data['price']< 1000]
sns.boxplot(data = df, x= 'price')
plt.show()
```

```python
sns.histplot(data=df, x='price', bins = 80)
plt.show()
```

```python
sns.histplot(data=df, x='availability_365')
plt.title('365 D Availability')
plt.show()
```

```python
df.groupby(by = 'neighbourhood_group')['price'].mean()

```

```python
df['price_per_bed'] = df['price']/df['beds']
df.head()
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
sns.scatterplot(data=df, x='number_of_reviews', y='price',hue='room_type')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue = 'room_type')
plt.show()
```

```python
sns.scatterplot(data=df, x= 'longitude', y = 'latitude', hue= 'room_type')
plt.show()
```

```python
corr = df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
plt.figure(figsize=(10,8))
plt.show()
```
